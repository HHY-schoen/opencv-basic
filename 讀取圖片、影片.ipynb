```python
import cv2

#讀取圖片
from google.colab.patches import cv2_imshow

img = cv2.imread('colorcolor.jpg')

# img = cv2.resize(img, (200,300))
img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)  #x,y各縮小一半
cv2_imshow(img)
# cv2.waitKey(5000)  #5000毫秒= 5秒鐘、0= 一直顯示
```

```python
#讀取影片
import cv2
cap = cv2.VideoCapture('thumb.mp4')
# cap = cv2.VideoCapture('填入視訊盡頭畫面')

while True:
  ret, frame = cap.read()  #ret=影片的下一帪(布林值), frame=若有成功會回傳取得的圖片
  if ret:
    frame = cv2.resize(frame, (0,0), fx=0.4, fy=0.4)
    cv2_imshow(frame)
  else:
    break
  if cv2.waitKey(1) == ord('q'):  #按鍵盤上的Q:影片結束
    break
```
