## 常用函式

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('colorcolor.jpg')
img = cv2.resize(img, (0,0), fx=0.3, fy=0.3)

cv2_imshow(img)
```

- 色彩轉換 cvtColor

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

- 高斯模糊 GaussianBlur

```python
blur = cv2.GaussianBlur(img, (7,7), 5)  #(img,核的大小n*n(奇數),標準差)
cv2_imshow(blur)
```

- 邊緣檢測 Canny
- 像素質差別很大，看起來就會像邊緣。每個像素點會打一個分數(跟周圍像素的差異大小)，差別越大分數越高。
- ( 如下為低於200就過濾掉，高於250就當成邊緣。 )

```python
canny = cv2.Canny(img, 200, 250)  #(img,最低門檻值,最高門檻值)
cv2_imshow(canny)
```

- 膨脹 dilate

```python
import numpy as np
kernel = np.ones((5,5))

dilate = cv2.dilate(canny, kernel, iterations=1)  #(目標, 核大小(須為二維陣列), 要膨脹幾次)
cv2_imshow(dilate)
```

- 侵蝕 erode

```python
kernel1 = np.ones((5,5))

erode = cv2.erode(dilate, kernel1, iterations=1)  #(目標, 核大小(須為二維陣列), 要侵蝕幾次)
cv2_imshow(erode)
```

## 畫圖型

- 畫直線

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img = np.zeros((400,400,3))  #都是0，所以是全黑的圖片
cv2.line(img,(0,0),(img.shape[1],img.shape[0]),(0,255,0),2)  #(img,起始點,終點,線的顏色,線的粗度)。img.shape[1]可以得到寬度

cv2_imshow(img)
```

- 畫方形

```python
cv2.rectangle(img,(0,0),(300,300),(0,0,255),cv2.FILLED)  #(img,左上點,右上點,粗度or填滿)
cv2_imshow(img)
```

- 畫圓形

```python
cv2.circle(img,(200,200),30,(255,100,0),4)  #(img,中心點,半徑,顏色,線條粗度)
cv2_imshow(img)
```

- 寫文字
- putText 不支援寫中文

```python
cv2.putText(img,'Hello',(300,100),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)  #(img,'文字',座標,字體,大小,顏色,線條粗度)
cv2_imshow(img)
```
