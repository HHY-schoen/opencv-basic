### 讀取圖片

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('colorcolor.jpg')

# img = cv2.resize(img, (200,300))
img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)  #x,y各縮小一半
cv2_imshow(img)
# cv2.waitKey(5000)  #5000毫秒= 5秒鐘、0= 一直顯示
```

### 讀取影片

```python
import cv2
cap = cv2.VideoCapture('thumb.mp4')
# cap = cv2.VideoCapture('填入視訊盡頭畫面')

while True:
  ret, frame = cap.read()  #ret=影片的下一帪(布林值), frame=若有成功會回傳取得的圖片
  if ret:
    frame = cv2.resize(frame, (0,0), fx=0.4, fy=0.4)
    cv2_imshow(frame)
  else:
    break
  if cv2.waitKey(1) == ord('q'):  #按鍵盤上的Q:影片結束
    break
```

### numpy 與圖片的關係
- 一般排列為 (R,G,B)，但在 opancv 中排列順序為 (B,G,R)
- 對電腦來說，圖片其實就是一堆數字

```python
import cv2

img = cv2.imread('colorcolor.jpg')
# print(img)
print(type(img))  #ndarray: nd多維、array陣列
print(img.shape)
```

```python
#自己建立圖片
import numpy as np
from google.colab.patches import cv2_imshow
import random

img = np.empty((300,300,3))  #電腦為二進制，用8個bits來表示0~255。unit表示正整數

for row in range(300):
  for col in range(300):
    img[row][col] = [random.randint(0,255),random.randint(0,255),random.randint(0,255)]

cv2_imshow(img)
# cv2.waitKey(0)
```

```python
img = cv2.imread('colorcolor.jpg')
# print(img.shape)

for row in range(300):
  for col in range(img.shape[1]):
    img[row][col] = [random.randint(0,255),random.randint(0,255),random.randint(0,255)]

cv2_imshow(img)
```

```python
#切割圖片
img = cv2.imread('colorcolor.jpg')
new_img = img[400:600, 100:400]

cv2_imshow(new_img)
```
